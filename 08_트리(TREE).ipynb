```
# 이진트리_순환 in 트리

class BinaryNode:
  def __init__(self, data, left, right):
      self.data = data
      self.left = left   # 왼쪽 자식 
      self.right = right # 오른쪽 자식

  def isLeaf(self):      # 단말노드 인지 확인 T or F
      if self.left == None and self.right == None:
        return True
      else: return False

class BinaryTree:
  def __init__(self):   # root 초기 값 Noe
    self.root = None     
 
  def setRoot(self,r):   # Root 설정
    self.root = r          

  def getCount(self, root): # 전체 노드 갯수 count
    if root == None:        
      return 0
    else:
      return self.getCount(root.left) + self.getCount(root.right) + 1  
      # 왼쪽 + 오른쪽 + 루트까지 카운트  

  def lineCount(self, root):
    return self.getCount(root) - 1

  def getLeafCount(self, root): # 단말 노드 갯수 count
    if root == None:
      return 0
    if root.isLeaf(): # 단말 노드면 1 return
      return 1
    else:
      return self.getLeafCount(root.left) + self.getLeafCount(root.right)

  def getHeight(self, root):
    if root == None:
      return 0
    else:
      return max(self.getHeight(root.left), self.getHeight(root.right))+1 
      # 왼쪽, 오른쪽 비교해서 루트까지 높이

  def inorder(self, root):   # 중위 순회             
    if root != None:
      self.inorder(root.left)                   # 왼쪽을 다 돌고
      print("[{}]".format(root.data), end =" ") # 루트 거치고
      self.inorder(root.right)                  # 오른쪽 돈다.

  def preorder(self, root): # 전위 순회
    if root != None:
      print("[{}]".format(root.data), end =" ") # 루트부터 시작
      self.preorder(root.left)                  # 왼쪽 다 돌고
      self.preorder(root.right)                 # 오른쪽 돈다



  def postorder(self, root): # 후위 순회
    if root != None:
      self.postorder(root.left) # 왼쪽 다 돌고
      self.postorder(root.right) # 오른쪽 다 돌고
      print("[{}]".format(root.data), end =" ") # 루트 돈다.

  def levelorder(self, root): # 레벨 순회
    q = []
    q.append(root)            # 루트부터 시작. 선입 선출
    while q:                  # q가 있을 때 까지
      t = q.pop(0)            # t = root를 빼서
      if t != None:           # 만약 root가 None이 아니면, 
        print("[{}]".format(t.data), end=" ") # 루트의 데이터를 프린트
        q.append(t.left)                      # 왼쪽을 append 
        q.append(t.right)                     # 오른쪽 append


tree = BinaryTree()

h = BinaryNode('H', None, None )
g = BinaryNode('G', None, None )
d = BinaryNode('D', None, None )
e = BinaryNode('E', g, h )
b = BinaryNode('B', d, None )
f = BinaryNode('F', None, None )
c = BinaryNode('C', e, f )
a = BinaryNode('A', b, c )

tree.setRoot(a)

print("노드의 갯수 =", tree.getCount(tree.root))
print("간선의 갯수 =", tree.lineCount(tree.root))
print("단말의 갯수 =", tree.getLeafCount(tree.root))
print("트리의 높이 =", tree.getHeight(tree.root))

print("In-Order :    ", end=" ")
tree.inorder(tree.root)
print("\nPre-Order :   ", end=" ")
tree.preorder(tree.root)
print("\nPost-Order :  ", end=" ")
tree.postorder(tree.root)
print("\nLevel-Order : ", end=" ")
tree.levelorder(tree.root)

#       A
#     ↙ ↘
#    B      C
#  ↙    ↙  ↘
# D     E      F
#    ↙  ↘
#   G     H
```

```
# 이진트리_순환 in 트리 with 중첩 함수

class BinaryNode:
  def __init__(self, data, left, right):
      self.data = data
      self.left = left
      self.right = right

  def isLeaf(self):
      if self.left == None and self.right == None:
        return True
      else: return False

class BinaryTree:
  def __init__(self):
    self.root = None

  def setRoot(self,r):
    self.root = r

  def getCount(self):
    def getCount_in(root):
      if root == None:
        return 0
      else:
        return getCount_in(root.left) + getCount_in(root.right) + 1 
    root = self.root
    return getCount_in(root)

  def getLeafCount(self):
    def getLeafCount_in(root):
      if root == None:
        return 0
      if root.isLeaf():
        return 1
      else:
        return getLeafCount_in(root.left) + getLeafCount_in(root.right)
    root = self.root
    return getLeafCount_in(root)

  def getHeight(self):
    def getHeight_in(root):
      if root == None:
        return 0
      else:
        return max(getHeight_in(root.left), getHeight_in(root.right))+1 
    root = self.root
    return getHeight_in(root)  

  def preorder(self):
    def preorder_in(root):
      if root != None:
        print("[{}]".format(root.data), end =" ")
        preorder_in(root.left)
        preorder_in(root.right)
    print("\npreorder :   ", end=" ")
    preorder_in(self.root)

  def inorder(self):
    def inorder_in(root):
      if root != None:
        inorder_in(root.left)
        print("[{}]".format(root.data), end =" ")
        inorder_in(root.right)
    print("inorder :    ", end=" ")
    inorder_in(self.root)

  def postorder(self):
    def postorder_in(root):
      if root != None:
        postorder_in(root.left)
        postorder_in(root.right)
        print("[{}]".format(root.data), end =" ")
    print("\npostorder :  ", end=" ")
    postorder_in(self.root)

  def levelorder(self):
    print("\nlevelorder : ", end=" ")
    q = []
    q.append(self.root)
    while q:
      t = q.pop(0)
      if t != None:
        print("[{}]".format(t.data), end=" ")
        q.append(t.left)
        q.append(t.right)


tree = BinaryTree()

h = BinaryNode('H', None, None )
g = BinaryNode('G', None, None )
d = BinaryNode('D', None, None )
e = BinaryNode('E', g, h )
b = BinaryNode('B', d, None )
f = BinaryNode('F', None, None )
c = BinaryNode('C', e, f )
a = BinaryNode('A', b, c )

tree.setRoot(a)

print("노드의 갯수 =", tree.getCount())
print("단말의 갯수 =", tree.getLeafCount())
print("트리의 높이 =", tree.getHeight())

tree.inorder()
tree.preorder()
tree.postorder()
tree.levelorder()
```

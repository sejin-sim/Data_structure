```
# 인접행렬 그래프

global Max
Max = 256

class AdjMatGraph:
  def __init__(self): # 정점과 정점의 관계를 행렬로
    self.adj = [[0 for j in range(Max)] for i in range(Max)] 
    self.vertices = []                                       

  def size(self): # 정점 갯수
    return len(self.vertices)
  
  def reset(self): # 초기화
    self.adj = [[0 for j in range(Max)] for i in range(Max)]
    
  def setEdge(self, i, j, val): # 간선 셋팅
    self.adj[i][j] = val

  def insertVertex(self, V):  # 정점 삽입
    self.vertices.append(V)
  
  def insertEdge(self, u, v): # 간선 삽입
    self.setEdge(u, v, 1) 
    self.setEdge(v, u, 1) # 방향 그래프에서는 삭제됨

  def getVertex(self, i): # 위치로 정점(문자) 가져오기
    return self.vertices[i]
  
  def getEdge(self, i, j): # 간선의 값(숫자 0 or 1) 가져오기
    return self.adj[i][j]

  def display(self):
    print(self.size())
    for i in range(0, self.size()):
      print(self.getVertex(i), end=" ")

      for j in range(0, self.size()):
        print(self.getEdge(i, j), end=" ")
      print()
  
g = AdjMatGraph()
for i in range(0, 4):
  g.insertVertex(chr(65+i))

g.insertEdge(0,1)
g.insertEdge(0,3)
g.insertEdge(1,2)
g.insertEdge(1,3)
g.insertEdge(2,3)
print("인접 행렬로 표현한 그래프")
g.display()
```

```
# 인접리스트 그래프

global Max
Max = 256

class Node:
  def __init__(self, id, link):
    self.id = id
    self.link = link

class AdjListGraph:
  def __init__(self): # 정점과 정점의 관계
    self.vertices = [] 
    self.adj = [0 for j in range(Max)]

  def size(self):
    return len(self.vertices)

  def insertVertex(self, V):
    self.vertices.append(V)
  
  def insertEdge(self, u, v):          # 간선삽입 ex) 0, 1
    self.adj[u] = Node(v, self.adj[u]) # self.adj[0] = Node(1, self.adj[0])
    self.adj[v] = Node(u, self.adj[v]) # self.adj[1] = Node(0, self.adj[1])

  def getVertex(self, i):
    return self.vertices[i]
  
  def display(self):
    print(self.size())

    for i in range(0, self.size()):
      print(self.getVertex(i), end=" ")

      v = self.adj[i]
      while v != 0:
        print(self.getVertex(v.id), end=" ")
        v = v.link
      print()

g = AdjListGraph()
for i in range(0, 4):
  g.insertVertex(chr(65+i))
g.insertEdge(0,1)
g.insertEdge(0,3)
g.insertEdge(1,2)
g.insertEdge(1,3)
g.insertEdge(2,3)
print("인접 리스트로 표현한 그래프")
g.display()
```

```
# DFS(신장 트리) & BFS with 인접행렬

global Max
Max = 256

# 큐(queue) 기본(basic)_원형
class CircularQueue:
  def __init__(self):
      self.rear = 0
      self.front = 0
      self.data = []
      self.MaxSize = 100

  def size(self):
      return len(self.data)

  def isFull(self):
      return self.size() == self.MaxSize

  def isEmpty(self):
      return self.size() == 0

  def enqueue(self, item):
      if self.isFull():
        print("error: 큐가 포화상태입니다.")
      else:
        self.data.append(item)
        self.rear = (self.rear +1) % self.MaxSize
 
  def dequeue(self):
     if self.isEmpty():
       print("error: 큐가 공백상태입니다.")
     else:
       self.front = (self.front +1) % self.MaxSize
       return self.data.pop(0)

class SrchAMGraph(AdjMatGraph, CircularQueue):
  def __init__(self):
    self.adj = [[0 for j in range(Max)] for i in range(Max)]
    self.vertices = []
    self.visited = [0 for j in range(Max)]
    self.finished = [0 for j in range(Max)]
    self.haveCycle = False

  def resetVisited(self):
    for i in range(0, self.size()):
      self.visited[i] = False

  def isLinked(self, u, v):
    if self.getEdge(u, v) != 0:
      return True
    else:
      return False

  def DFS(self, v):
    self.visited[v] = True
    print(self.getVertex(v), end = " ")
    for w in range(self.size()):
      if self.isLinked(v, w) and self.visited[w] == False:
        self.DFS(w)

  def spanningTreeByDFS(self, v):
    self.visited[v] = True
    for w in range(self.size()):
      if self.isLinked(v, w):
        if self.visited[w] != True:
          print("신장트리간선 :", self.getVertex(v),"-", self.getVertex(w))
          self.spanningTreeByDFS(w)
        elif self.finished[w] != True:
          self.haveCycle = True
    self.finished[v] = True

  def BFS(self, v):
    self.visited[v] = True
    print(self.getVertex(v), end = " ")
    que = CircularQueue()
    que.enqueue(v)
    while que.isEmpty() != True :
      v = que.dequeue()
      for w in range(0, self.size()):
        if self.isLinked(v,w) and self.visited[w] == False:
          self.visited[w] = True
          print(self.getVertex(w), end = " ")
          que.enqueue(w)

  def load(self, filename):
    fp = open(filename, "r")
    if( fp != None ):
      n = int(fp.readline())
      for i in range(0, n):
        str = list(fp.readline().split())
        self.insertVertex(str[0])
        
        for j in range(0, n):
          if int(str[j+1]) == 1:
            self.insertEdge(i,j)	

    fp.close()


g = SrchAMGraph()
g.load("/content/drive/MyDrive/자료구조/업로드/graph.txt")
print("그래프(graph.txt)\n")
g.display()
      
print("\nDFS ==> ", end="")
g.resetVisited()
g.DFS(0)

print("\n\nSpanningTreeByDFS")
g.resetVisited()
g.spanningTreeByDFS(0)
print("사이클이 있나요?", g.haveCycle)
print("\nBFS ==> ", end="")
g.resetVisited()
g.BFS(0)
```

```
# 브리지 찾기 with 인접 행렬

global Max
Max = 256

class ConnectedComponentGraph(SrchAMGraph):
  def __init__(self):
    self.adj = [[0 for j in range(Max)] for i in range(Max)]
    self.label = [0 for j in range(Max)]
    self.vertices = []
    self.visited = [0 for j in range(Max)]

  def labelDFS(self, v, color):
    self.visited[v] = True
    self.label[v] = color

    for w in range(0, self.size()):
      if self.isLinked(v, w) and self.visited[w] == False:
        self.labelDFS(w, color)

  def findConnectedComponent(self):
    count = 0
    for i in range(self.size()):
      if self.visited[i] == False:
        count += 1
        self.labelDFS(i, count)

    return count

def findBridge(g):
	count = 0
	g.resetVisited()

	for i in range(0, g.size()):
		for j in range(0, i):
			if g.getEdge(i, j) != 0:
				g.setEdge(i, j, 0)
				g.setEdge(j, i, 0)
		
				g.resetVisited()
				if g.findConnectedComponent() > 1:
					print("Bridge{} : {}-{}".format(count+1, g.getVertex(i), g.getVertex(j)))
					count += 1
				g.setEdge(i, j, 1)
				g.setEdge(j, i, 1)
		
	return count

g = ConnectedComponentGraph()

g.load("/content/drive/MyDrive/자료구조/업로드/graph.txt")
print("그래프(graph.txt)\n")
g.display()
g.resetVisited()
g.findConnectedComponent()
print("Bridge :", findBridge(g))
```

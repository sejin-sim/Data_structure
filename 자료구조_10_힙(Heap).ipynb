```
# Maxheap with Node
# C10_1

class HeapNode:
  def __init__(self, k):
    self.key = k

  def getKey(self):
    return self.key
  
  def display(self):
    print("\t{} ".format(self.key), end = "")

class MaxHeap:
  def __init__(self):
    self.node = []
    self.size = 0

  def isEmpty(self):
    if self.size ==0: return True
    else: return False

  def isFull(self):
    if self.size == 200-1: return True
    else: return False

  def parent(self, i):
      if i & 1:
          return i >> 1
      else:
          return (i >> 1) - 1

  def left_child(self, i):
      return (i << 1) + 1

  def right_child(self, i):
      return (i << 1) + 2

  def getParent(self, i):
    i = int(i)
    return self.node[i//2]

  def insert(self, k):
    self.node.append(HeapNode(k))          # 노드 배열에 key 삽입
    last_index = len(self.node) -1         # 끝 인덱스 넘버
    
    while 0 <= last_index:                 # 끝 인덱스 넘버가 0보다 작거나 같아질 때까지
      parent_index = (last_index-1)//2     # 아빠인덱스 = (끝인덱스-1)//2
      if 0<= parent_index and self.node[parent_index].key < self.node[last_index].key:
        self.node[last_index], self.node[parent_index] = self.node[parent_index], self.node[last_index]
        last_index = parent_index
      else:
        break

  def remove(self):
    last_index = len(self.node) -1
      # 하나이상의 노드가 존재하는 경우
    if last_index < 0:
      return -1

  # 맨마지막 원소와 위치바꾸기
    self.swap(0, last_index)
    maxv = self.node.pop()
    self.maxHeapify(0)
    return maxv

  def swap(self, i, parent_index):
    self.node[i], self.node[parent_index] = self.node[parent_index], self.node[i]

  def maxHeapify(self, i):
      # i는 어느 기준으로 바꿀건가
      
      # 왼쪽 자식 (left child) 의 인덱스를 계산합니다.
      left = i * 2 + 1
      # 오른쪽 자식 (right child) 의 인덱스를 계산합니다.
      right = i * 2 + 2
      greatest = i
      
      # 자신(i), 왼쪽자식(left), 오른쪽자식(right) 중 최대를 찾아서 이것의 인덱스를 smallest에 담는다
      # 왼쪽 자식이 존재하는지, 그리고 왼쪽 자식의 (키) 값이 (무엇보다?) 더 큰지를 판단합니다.
      if left <= len(self.node)-1 and self.node[left].key > self.node[greatest].key:
          # 조건이 만족하는 경우, smallest 는 왼쪽 자식의 인덱스를 가집니다.
          greatest = left

      # 오른쪽 자식이 존재하는지, 그리고 오른쪽 자식의 (키) 값이 (무엇보다?) 더 큰지를 판단합니다.
      if right <= len(self.node)-1 and self.node[right].key > self.node[greatest].key:
          # 조건이 만족하는 경우, smallest 는 오른쪽 자식의 인덱스를 가집니다.
          greatest = right
          
      # 만약 이 인덱스가 i와 같지 않다면 자식들 중에 나보다 큰 키값을 가진 노드가 발견된 경우
      if greatest != i:
          # 현재 노드 (인덱스 i) 와 최댓값 노드 (왼쪽 아니면 오른쪽 자식) 를 교체합니다.
          self.swap(i, greatest)
          # 재귀적 호출을 이용하여 최대 힙의 성질을 만족할 때까지 트리를 정리합니다.
          self.maxHeapify(greatest)

  def display(self):
    level = 1
    for i in range(0, len(self.node)):
      if i == level -1:
        print()
        level *= 2
      self.node[i].display()
    print("\n------------------------------------------------")


  def heapSort(self, a):
    for i in range(0, len(a)):
      self.insert(a[i])
    for i in range(len(a)-1, -1, -1):
      a[i] = self.remove().key

    return a

  def heapSort_down(self, a):
    for i in range(0, len(a)):
      self.insert(a[i])
    for i in range(0, len(a)):
      a[i] = self.remove().key

    return a


heap = MaxHeap()
heap.insert(10)
heap.insert( 5)
heap.insert(30)
heap.insert( 8)
heap.insert( 9)
heap.insert( 3)
heap.insert( 7)
heap.display( )

heap.remove()
heap.display()
heap.remove()
heap.display()

heap2 = MaxHeap()
data = [41, 67, 34, 0, 69, 24, 78, 58, 62, 64]
print("정렬 전       : {}".format(data))
heap2.heapSort(data)
print("정렬 후(오름) : {}".format(data))

heap3 = MaxHeap()
data_2 = [41, 67, 34, 0, 69, 24, 78, 58, 62, 64]
heap2.heapSort_down(data_2)
print("정렬 후(내림) : {}".format(data_2))
```

```
# Heap module
# C10_2

import heapq

nums = [4, 1, 7, 3, 8, 5]
heap = []

# 최대 힙(기본이 최소힙)
for num in nums:
  heapq.heappush(heap, (-num, num))  # (우선 순위, 값)

while heap:
  print(heapq.heappop(heap)[1], end=" ")  # index 1
print()

# 힙정렬
def heap_sort(nums):   # 오름차순
  heap = []
  for num in nums:
    heapq.heappush(heap, num)
  
  sorted_nums = []
  while heap:
    sorted_nums.append(heapq.heappop(heap))
  return sorted_nums

def heap_sort_down(nums):# 내림차순
  heap = []
  for num in nums:
    heapq.heappush(heap, -num)
  
  sorted_nums = []
  while heap:
    sorted_nums.append(-heapq.heappop(heap))
  return sorted_nums

print("정렬 전       :",[41, 67, 34, 0, 69, 24, 78, 58, 62, 64])
print("정렬 후(오름) :",heap_sort([41, 67, 34, 0, 69, 24, 78, 58, 62, 64]))
print("정렬 후(내림) :",heap_sort_down([41, 67, 34, 0, 69, 24, 78, 58, 62, 64]))
```

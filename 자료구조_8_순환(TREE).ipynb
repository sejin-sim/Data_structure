```
# 이진트리_순환 in 트리

class BinaryNode:
  def __init__(self, data, left, right):
      self.data = data
      self.left = left
      self.right = right

  def isLeaf(self):
      if self.left == None and self.right == None:
        return True
      else: return False

class BinaryTree:
  def __init__(self):
    self.root = None

  def setRoot(self,r):
    self.root = r

  def getCount(self, root):
    if root == None:
      return 0
    else:
      return self.getCount(root.left) + self.getCount(root.right) + 1 

  def getLeafCount(self, root):
    if root == None:
      return 0
    if root.isLeaf():
      return 1
    else:
      return self.getLeafCount(root.left) + self.getLeafCount(root.right)

  def getHeight(self, root):
    if root == None:
      return 0
    else:
      return max(self.getHeight(root.left), self.getHeight(root.right))+1 

  def preorder(self, root):
    if root != None:
      print("[{}]".format(root.data), end =" ")
      self.preorder(root.left)
      self.preorder(root.right)

  def inorder(self, root):
    if root != None:
      self.inorder(root.left)
      print("[{}]".format(root.data), end =" ")
      self.inorder(root.right)

  def postorder(self, root):
    if root != None:
      self.postorder(root.left)
      self.postorder(root.right)
      print("[{}]".format(root.data), end =" ")

  def levelorder(self, root):
    q = []
    q.append(root)
    while q:
      t = q.pop(0)
      if t != None:
        print("[{}]".format(t.data), end=" ")
        q.append(t.left)
        q.append(t.right)


tree = BinaryTree()

d = BinaryNode('D', None, None )
e = BinaryNode('E', None, None )
b = BinaryNode('B', d, e )
f = BinaryNode('F', None, None )
c = BinaryNode('C', f, None )
a = BinaryNode('A', b, c )

tree.setRoot(a)

print("노드의 갯수 =", tree.getCount(tree.root))
print("단말의 갯수 =", tree.getLeafCount(tree.root))
print("트리의 높이 =", tree.getHeight(tree.root))

print("inorder :    ", end=" ")
tree.inorder(tree.root)
print("\npreorder :   ", end=" ")
tree.preorder(tree.root)
print("\npostorder :  ", end=" ")
tree.postorder(tree.root)
print("\nlevelorder : ", end=" ")
tree.levelorder(tree.root)
```
